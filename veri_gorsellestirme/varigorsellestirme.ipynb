```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv('bilet_fiyat.csv')#csv dosyamızı okuyor
dataset.head()#verilerimizin baştan ilk beş satırını döndürür.
```

```python
plt.figure(figsize=(8,6))
plt.plot(dataset.Yıl,dataset.Fiyatı)#yıllara göre bilet fiyatları
plt.title("2017-2022 yılları arasındaki İstanbul otobüs bilet fiyatları")
plt.xlabel("Yıllar")
plt.ylabel("Bilet Fiyatı")
```

```python
plt.figure(figsize=(8,6))
plt.plot(dataset.Bilet_Adı,dataset.Fiyatı)
plt.title("2017-2022 yılları arasındaki İstanbul'daki bilet sınıfına göre bilet fiyatları")
plt.xlabel("Bilet Sınıfı")
plt.ylabel("Bilet Fiyatı")
```

```python
sns.countplot(x='Bilet_Adı' ,data=dataset, )
plt.title("Bilet Sınıfına ait Çubuk Grafiği")
```

```python
sns.jointplot(x=dataset['Bilet_Adı'],y=dataset['Fiyatı']  )
```

```python
dataset=pd.read_csv('2021_deniz_seferleri.csv' )#csv dosyamızı okuyor 
dataset.head()#verilerimizin baştan ilk beş satırını döndürür.
```

```python
plt.figure(figsize=(8,6))
plt.plot(dataset.ISLETME_TURU,dataset.GEMI_SAYISI)#yıllara göre bilet fiyatları
plt.title("İstabul'da bulunan Deniz işletmelerinin türlerine göre Gemi sayısı")

plt.xlabel("GEMİ SAYISI")
plt.ylabel("İŞLETME TÜRÜ")
```
